```python
from IPython.display import display

from rake_nltk import Rake
# Initialize the Rake keyword extraction object
r = Rake()

# Extract keywords from the text in notes.pn_history[20] after removing parentheses
r.extract_keywords_from_text(notes.pn_history[20].replace('(','').replace(')',''))

# Extract keywords from the text in notes.pn_history[20] (no modification)
r.extract_keywords_from_text(notes.pn_history[20])

# Create a DataFrame to store ranked phrases with their scores
phrase_df = pd.DataFrame(r.get_ranked_phrases_with_scores(), columns=['score', 'phrase'])

# Filter and retain only phrases with a score greater than 3
filtered_phrase_df = phrase_df.loc[phrase_df['score'] > 3]

display(filtered_phrase_df)
```
